```python
data=df.drop(columns=['Row ID','Order Date', 'Ship Date','Customer ID', 'Customer Name','Product ID'],axis=1)
data.head()
```

```python
import plotly.express as px
fig = px.imshow(data.corr(),text_auto=True)
fig.show()
```

**TOP 10 STATES WHICH HAS MORE PROFIT**

```python
data["State"].unique()
```

```python
state_profit=pd.DataFrame(data.groupby("State")["Profit"].sum())
state_profit.reset_index(inplace=True)
```

```python
state_profit=state_profit.sort_values(by="Profit",ascending=False)
state_profit.head(5)
```

**PROFITABLE SUB_CATEGORY**

```python
data["Sub-Category"].unique()
```

```python
sub_categories_info=pd.DataFrame(data.groupby("Sub-Category")["Profit"].sum())
sub_categories_info.reset_index(inplace=True)
sub_categories_profit=sub_categories_info.sort_values(by="Profit",ascending=False)
sub_categories_profit.head(7)
```

```python
plt.bar(sub_categories_profit["Sub-Category"],sub_categories_profit["Profit"])
plt.xticks(sub_categories_profit['Sub-Category'] , rotation = "vertical")
plt.xlabel("Sub-Categories")
plt.ylabel("Profit")
plt.show()
```

**CATEGORY THAT SOLD MOST**

```python
data["Category"].value_counts()
```

```python
categories_sales=data.groupby("Category").sum()
categories_sales.reset_index(inplace=True)
categories_sales.head(10)
```

**HIGHEST SALE**

**MOST POPULAR SUB_CATEGORY**

```python
psb=pd.DataFrame(data.groupby("Sub-Category")["Quantity"].sum())
psb.reset_index(inplace=True)
psb=psb.sort_values(by="Quantity",ascending=False)
```

```python
plt.bar(psb["Sub-Category"],psb["Quantity"],color="red")
plt.xticks(psb["Sub-Category"],rotation="vertical")
plt.title("Sub-Category Vs Quantity")
plt.xlabel("Sub Category")
plt.ylabel("Quantity")
plt.show()
```

**Most Profitable Customer Segment**

```python
cs=data.groupby("Segment").sum()
cs.reset_index(inplace=True)
cs
```

```python
plt.bar(cs["Segment"],cs["Profit"],color="green")
plt.xticks(cs["Segment"],rotation="vertical")
plt.title("Segment Vs Profit")
plt.xlabel("Segment")
plt.ylabel("Profit")
plt.show()
```

**Most Profitable Region**

```python
data["Region"].value_counts()
```

```python
profitable_region=pd.DataFrame(data.groupby("Region")["Profit"].sum())
profitable_region.reset_index(inplace=True)
profitable_region
```

**City has the highest Sales Volume**

```python
city_volume = pd.DataFrame(data.groupby('City')['Sales'].sum())
city_volume =city_volume.sort_values(by="Sales",ascending=False)
city_volume.head()
```

```python
plt.bar(city_volume.index[0:20], city_volume['Sales'][0:20],color="yellow")
plt.xticks(city_volume.index[0:20],rotation = 'vertical')
plt.xlabel("City" )
plt.ylabel("Sales")
plt.show()
```
